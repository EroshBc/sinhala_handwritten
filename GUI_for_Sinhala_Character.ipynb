# Create GUI using python ktinker

```python
import tkinter as tk
```

```python
width=600
height=600
font_btn = 'Helvetica 20 bold'

win = tk.Tk()

## create canvas where to write
canvas = tk.Canvas(win, width= width, height=height, bg ="white")
canvas.grid(row=0, column=0)

## create button frame
button_frame = tk.Frame(win)
button_frame.grid(row=1, column=0, pady=10)

##create buttons
button_save = tk.Label(button_frame, text="SAVE", bg="green", fg="white",
                       font=font_btn, padx=43, pady=15)


button_save.grid(row=0, column=0)

win.mainloop()
```
