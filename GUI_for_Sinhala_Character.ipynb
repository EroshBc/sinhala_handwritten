# Create GUI using python ktinker

```python
import tkinter as tk
from PIL import ImageTk,Image,ImageDraw
import numpy as np
import os
import cv2
```

```python
width=600
height=600
font_btn = 'Helvetica 20 bold'
font_label='Helvetica 22 bold'
count =0

win = tk.Tk()

## create function to user to write n canvas
## letter written as set of oveal shapes

def event_function(event):
    
    x=event.x
    y=event.y
    
    x1=x-30
    x2=x+30
    
    y1=y-30
    y2=y+30
    
    canvas.create_oval((x1,y1,x2,y2),fill='black')
    img_draw.ellipse((x1,y1,x2,y2),fill='black')

def save():
    
    global count
    
    img_array=np.array(img)
    path=os.path.join('data',str(count)+'.jpg')
    #path=data/0.jpg
    cv2.imwrite(path,img_array)
    count=count+1

def clear():
    
    global img,img_draw

    canvas.delete('all') # to clear the canvas
    
    img=Image.new('RGB',(width,height),(255,255,255))
    img_draw=ImageDraw.Draw(img)  
    
    label_status.config(text='PREDICTED CHARACTER: NONE')
    
    folder_path = "data"

    image_exts = (".png", ".jpg", ".jpeg", ".bmp", ".gif")

    if os.path.exists(folder_path):
        for file in os.listdir(folder_path):
            if file.lower().endswith(image_exts):
                file_path = os.path.join(folder_path, file)
                try:
                    os.remove(file_path)
                except Exception as e:
                    print(f"Could not delete {file_path}: {e}")
    
    

## create canvas where to write
canvas = tk.Canvas(win, width= width, height=height, bg ="white")
canvas.grid(row=0, column=0, columnspan=4)

## create button frame
button_frame = tk.Frame(win)
button_frame.grid(row=1, column=0, pady=10)

##create buttons
button_save = tk.Label(button_frame, text="SAVE", bg="green", fg="white",
                       font=font_btn, padx=43, pady=15, cursor = "hand2")

button_predict = tk.Label(button_frame, text="PREDICT", bg="blue", fg="white",
                          font=font_btn, padx=43, pady=15, cursor = "hand2")

button_clear = tk.Label(button_frame, text="CLEAR", bg="yellow", fg="white",
                        font=font_btn, padx=43, pady=15, cursor = "hand2")

button_exit = tk.Label(button_frame, text="EXIT", bg="red", fg="white",
                       font=font_btn, padx=43, pady=15,cursor = "hand2")

label_status = tk.Label(win,text="Predicted Character: NONE", bg="white",font=font_label )


button_save.grid(row=0, column=0)
button_predict.grid(row=0, column=1)
button_clear.grid(row=0, column=2)
button_exit.grid(row=0, column=3)
label_status.grid(row=2,column=0,columnspan=4)



canvas.bind('<B1-Motion>',event_function)
button_save.bind("<Button-1>", lambda e:save())
button_clear.bind("<Button-1>", lambda e:clear())

##draw a paralal image 
img=Image.new('RGB',(width,height),(255,255,255))
img_draw=ImageDraw.Draw(img)

win.mainloop()
```
